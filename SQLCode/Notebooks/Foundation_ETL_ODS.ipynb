### Create Base for Extract View with Columns Name

```sql
USE ODSUtility
GO
DECLARE
	-- Schema name of the object that contains the columns
	@SchemaName		NVARCHAR(128) = N'dbo'
	-- Object name of the object that contains the columns
	,@ObjectName	NVARCHAR(128) = N'vwSourceFactDDATransactions';

SELECT
	STUFF
		(
			CAST
				(
					(
						SELECT
							N',CAST([' + [C].[COLUMN_NAME] + N'] AS ' + UPPER([C].[DATA_TYPE])
							+ CASE
								  WHEN [C].[CHARACTER_MAXIMUM_LENGTH] IS NOT NULL
									   AND	 [C].[CHARACTER_MAXIMUM_LENGTH] > 0
									  THEN
									  N'(' + CAST([C].[CHARACTER_MAXIMUM_LENGTH] AS NVARCHAR(128)) + N')'
								  WHEN [C].[CHARACTER_MAXIMUM_LENGTH] IS NOT NULL
									   AND	 [C].[CHARACTER_MAXIMUM_LENGTH] = -1
									  THEN
									  N'(MAX)'
								  WHEN [C].[DATETIME_PRECISION] IS NOT NULL
									   AND	 [C].[DATA_TYPE] NOT IN
									  (
										  N'datetime'
										 ,N'date'
										 ,N'smalldatetime'
									  )
									  THEN
									  N'(' + CAST([C].[DATETIME_PRECISION] AS NVARCHAR(128)) + N')'
								  WHEN [C].[DATA_TYPE] IN
									  (
										  N'DECIMAL'
										 ,N'NUMERIC'
									  )
									  THEN
									  N'(' + CAST([C].[NUMERIC_PRECISION] AS NVARCHAR(128)) + N',' + CAST([C].[NUMERIC_SCALE] AS NVARCHAR(128))
									  + N')'
								  ELSE
									  N''
							  END + ') AS ' + QUOTENAME([C].[COLUMN_NAME], N'[') AS [text()]
						FROM	[INFORMATION_SCHEMA].[COLUMNS] AS [C]
						WHERE
							[C].[TABLE_SCHEMA] = @SchemaName
							AND [C].[TABLE_NAME] = @ObjectName
						ORDER BY
							[C].[COLUMN_NAME]
						FOR XML PATH(N''), TYPE
					)
					AS NVARCHAR(MAX)
				)
			,1
			,1
			,N''
		);
        --REPLACE(REPLACE(@[System::PackageName], "Master", "Stage"), " ", ".") + ".dtsx"
```
